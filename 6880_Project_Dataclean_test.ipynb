```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from sklearn.model_selection import train_test_split
```

```python
# Load the training set.
train_df = pd.read_csv("train.csv")
```

```python
# Load the test set.
test_df = pd.read_csv("test.csv")
```

```python
# Picked 9 most important features from the heat map
new_train = train_df[['OverallQual','GrLivArea','GarageCars','TotalBsmtSF', 'FullBath', 'YearBuilt','YearRemodAdd',
                      'GarageYrBlt','MasVnrArea', 'SalePrice']]
new_test = test_df[['OverallQual','GrLivArea','GarageCars','TotalBsmtSF', 'FullBath', 'YearBuilt','YearRemodAdd',
                      'GarageYrBlt','MasVnrArea']]
new_train.head()
```

```python
# Count how many 'NA' data points in 'GarageYrBlt' feature 
sum(new_test['GarageYrBlt'].isnull())
```

```python
# Get mean value of GarageYrBlt 
avg_GarageYrBlt = np.mean(new_test['GarageYrBlt'])
```

```python
# Replace 'NA' data points with mean of whole data points in 'GarageYrBlt' feature
new_test['GarageYrBlt'] = new_test['GarageYrBlt'].fillna(avg_GarageYrBlt)
```

```python
# Count how many 'NA' data points in 'MasVnrArea' feature 
sum(new_test['MasVnrArea'].isnull())
```

```python
# Get mean value of MasVnrArea
avg_MasVnrArea = np.mean(new_test['MasVnrArea'])
```

```python
# Replace 'NA' data points with mean of whole data points in 'MasVnrArea' feature
new_test['MasVnrArea'] = new_test['MasVnrArea'].fillna(avg_MasVnrArea)
```

```python
# Detect any outlier : Quantil based flooring and capping method
testdata_nosalesprice = new_test
def detect_outliers_zscore(data):
    outliers = []
    thres = 2
    mean = np.mean(data)
    std = np.std(data)
    # print(mean, std)
    for i in data:
        z_score = (i-mean)/std
        if (np.abs(z_score) > thres):
            outliers.append(i)
    return outliers# Driver code
outliers_overallqual = detect_outliers_zscore(testdata_nosalesprice['OverallQual'])
outliers_GrLivArea = detect_outliers_zscore(testdata_nosalesprice['GrLivArea'])
outliers_GarageCars = detect_outliers_zscore(testdata_nosalesprice['GarageCars'])
outliers_TotalBsmtSF = detect_outliers_zscore(testdata_nosalesprice['TotalBsmtSF'])
outliers_FullBath = detect_outliers_zscore(testdata_nosalesprice['FullBath'])
outliers_YearBuilt = detect_outliers_zscore(testdata_nosalesprice['YearBuilt'])
outliers_YearRemodAdd = detect_outliers_zscore(testdata_nosalesprice['YearRemodAdd'])
outliers_GarageYrBlt = detect_outliers_zscore(testdata_nosalesprice['GarageYrBlt'])
outliers_MasVnrArea = detect_outliers_zscore(testdata_nosalesprice['MasVnrArea'])
```

```python
# Replace the outliers : Quantil based flooring and capping method
def replace_outlier(feature):
# Computing 10th, 90th percentiles and replacing the outliers
    tenth_percentile = np.percentile(testdata_nosalesprice[feature], 10)
    ninetieth_percentile = np.percentile(testdata_nosalesprice[feature], 90)
    print(tenth_percentile,ninetieth_percentile)
# print(tenth_percentile, ninetieth_percentile)
    b = np.where(testdata_nosalesprice[feature]<tenth_percentile, tenth_percentile, testdata_nosalesprice[feature])
    b = np.where(b>ninetieth_percentile, ninetieth_percentile, b)
    print(feature,b)
# print("Overallqual:", traindata_nosalesprice['OverallQual'])
    dic = {'A': b}
    df_replace = pd.DataFrame(dic)
    testdata_nosalesprice[feature] = df_replace['A'].values
    
# print("Overallqual:", traindata_nosalesprice['OverallQual'])
    return testdata_nosalesprice
replace_outlier('OverallQual') 
replace_outlier('GrLivArea')  
replace_outlier('GarageCars') 
replace_outlier('TotalBsmtSF') 
replace_outlier('FullBath')
replace_outlier('YearBuilt') 
replace_outlier('YearRemodAdd')
replace_outlier('GarageYrBlt') 
replace_outlier('MasVnrArea')
```
